```python
import nltk
import string
from string import digits
#stop word removal
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def text_preprocessing(text):
    doc_lowers = text.lower()
    remove_digits = str.maketrans('', '', digits)
    translator = str.maketrans('', '', string.punctuation)
    doc_no_punctuation = doc_lowers.translate(translator).translate(remove_digits)
    doc_tokens = nltk.word_tokenize(doc_no_punctuation) # Tokenize
    doc_stemmed = [] # stemmed
    doc_filtered = [w for w in doc_tokens if not w in stopwords.words("english")] # Stop words removal and stemming
    doc_stem = [ps.stem(w) for w in doc_filtered] 
    doc_stemmed.append(doc_stem)
    return doc_stemmed
```

```python
ready_doc_1 = text_preprocessing(doc_1)[0]
```

```python
ready_doc_1
```

```python
ready_doc_2 = text_preprocessing(doc_2)[0]
ready_doc_3 = text_preprocessing(doc_3)[0]
ready_doc_4 = text_preprocessing(doc_4)[0]
ready_doc_5 = text_preprocessing(doc_5)[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
all_list = [ready_doc_1, ready_doc_2, ready_doc_3, ready_doc_4, ready_doc_5]
```

```python
def tokenizer(text):
    return text
```

```python
tfidf = TfidfVectorizer(tokenizer = tokenizer, stop_words = 'english', lowercase = False, sublinear_tf = True, ngram_range=(1, 1))

tfs = tfidf.fit_transform(all_list)
print(len(tfidf.get_feature_names()))
print(tfs.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarities = cosine_similarity(tfs, tfs)
cosine_similarities
```
